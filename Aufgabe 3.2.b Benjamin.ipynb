```python
import numpy as np
import pandas as pd
```

```python
df_aisles = pd.read_csv('Data/aisles.csv')
df_departments = pd.read_csv('Data/departments.csv')
df_order_products = pd.read_csv('Data/oder_products_s.csv')
df_orders = pd.read_csv('Data/oders_s.csv')
df_products = pd.read_csv('Data/products.csv')
df_tips = pd.read_csv('Data/tips.csv')
```

```python
df_tips.drop(columns="Unnamed: 0",inplace=True)
df_product_tip = df_order_products.merge(on="order_id",right=df_tips)[["product_id","tip"]]
```

```python
df_product_tip_precentage = df_product_tip.groupby("product_id").count()
df_product_tip_precentage["tip given"] = df_product_tip.groupby("product_id").sum()
```

```python
df_product_tip_precentage["tip percentage"] = (df_product_tip_precentage["tip given"] / df_product_tip_precentage.tip)
```

```python
df_product_tip_precentage.rename(columns={"tip": "times ordered"},inplace=True)
```

```python
df_product_tip_precentage.drop(df_product_tip_precentage[df_product_tip_precentage["times ordered"] < 1000].index,axis=0,inplace=True)
```

```python
TOP10 = df_product_tip_precentage.sort_values(by="tip percentage",ascending=False).head(10)
```

```python
FLOP10 = df_product_tip_precentage.sort_values(by="tip percentage").head(10)
```

```python
TOP10 = TOP10.reset_index().merge(df_products)[["product_id","times ordered","tip given","tip percentage","product_name","department_id"]]
TOP10
```
