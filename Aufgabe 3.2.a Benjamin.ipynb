```python
import numpy as np
import pandas as pd
```

```python
df_aisles = pd.read_csv('Data/aisles.csv')
df_departments = pd.read_csv('Data/departments.csv')
df_order_products = pd.read_csv('Data/oder_products_s.csv')
df_orders = pd.read_csv('Data/oders_s.csv')
df_products = pd.read_csv('Data/products.csv')
df_tips = pd.read_csv('Data/tips.csv')
```

```python
df_tips.drop(columns="Unnamed: 0",inplace=True)
```

```python
df_product_tip = df_order_products.merge(on="order_id",right=df_tips)[["product_id","tip"]]
```

```python
products_department = df_products.merge(df_departments,on="department_id")[["product_id","department_id"]]
products_department
```

```python
df_product_tip = df_product_tip.merge(products_department,on="product_id")[["department_id","tip"]]
```

```python
df_product_tip.groupby("department_id").count()
```

```python
tip_percentage = df_product_tip.groupby("department_id").count()
```

```python
tip_percentage["tipgiven"] = df_product_tip.groupby("department_id").sum()
```

```python
tip_percentage["Tip percentage"] = (tip_percentage.tipgiven / tip_percentage.tip).round(2)
```

```python
tip_percentage = tip_percentage.rename(columns={"tip": "times ordered","tipgiven":"times tip given"}).sort_values(by="Tip percentage",ascending=False)
tip_percentage
```
